```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="/content/car_prices.csv"
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
df.info()
```

```python
df.shape
```

```python
#1
condition_vehicles=df.sort_values(by="condition",ascending=False).head(10)
condition_vehicles_name=condition_vehicles["make"].head(10)
Top_condition_vehicles=condition_vehicles["condition"]
plt.bar(condition_vehicles_name,Top_condition_vehicles,color="Red")
plt.xticks(rotation=90)
plt.xlabel("Car Brand")
plt.ylabel("Condition Rating")
plt.title("TOP 10 BRANDS WITH BEST CONDITION")
```

Insights:-
All cars have equal car condition


```python
#2
year_group=df.groupby("year")
market_value_2015=year_group.get_group(2015).sort_values(by="mmr",ascending=False)
market_value_cars_2015=market_value_2015["make"].head(10)
top_market_value_2015=market_value_2015["mmr"].head(10)
plt.plot(market_value_cars_2015,top_market_value_2015,"g*-",markerfacecolor="red",markersize=15)
plt.xlabel("Car Brand")
plt.ylabel("Market Value")
```

Insights:-
As per the 2015 car manufacturing data Mercedas-Benz has greatest marketvalue

```python
#3
cars_2014=year_group.get_group(2014).head(10)
car_colors_2014=cars_2014["color"]
car_names_2014=cars_2014["make"]
plt.scatter(car_names_2014,car_colors_2014)
plt.title("2014 manufacturing card and their colors")
plt.xlabel("Car Brand")
plt.ylabel("Car colors")
```

Insights:-
From this graph we can say that Chevrolet Brand provides more varients of colors,such as Black,Red,Blue

```python
#4
cars_2013=year_group.get_group(2013).sort_values(by="sellingprice",ascending=False)
Top_sellingprice_cars_2013=cars_2013["sellingprice"].head(10)
Top_sellingprice_car_names_2013=cars_2013["make"].head(10)
plt.bar(Top_sellingprice_car_names_2013,Top_sellingprice_cars_2013,color="green")
plt.title("2013 manufacturing cars and their price")
plt.xlabel("Car Brand")
plt.ylabel("Selling Price")
```

Insights:-
 As per the 2013 car manufacturing data Bently has greatest selling price

```python
#5
plt.title("Transmission distribution")
vehicle_gears=df["transmission"].value_counts()
wedgeprops={"edgecolor":"White","linewidth":5}
plt.pie(vehicle_gears,labels=vehicle_gears.index,colors=["red","green"],wedgeprops=wedgeprops,autopct="%1.1f%%")
plt.legend(bbox_to_anchor=(1.5,1))
```

Insights:-
As per the graph most of the cars are used automatic gears

```python
#6
car_group=df.groupby("make")
bmw_cars=car_group.get_group("BMW").head(10)
bmw_car_colors=bmw_cars["color"].value_counts()
plt.pie(bmw_car_colors,labels=bmw_car_colors.index,autopct="%1.1f%%")
plt.title("BMW car colors")
plt.legend(bbox_to_anchor=(1.5,1))

```

```python
#7
kia_cars=car_group.get_group("Kia").head(10)
kia_cars_interior_color=kia_cars["interior"].value_counts()
plt.pie(kia_cars_interior_color,labels=kia_cars_interior_color.index,autopct="%1.1f%%")
plt.title("Kia interior colors")
plt.legend(bbox_to_anchor=(1.5,1))
```

```python
#11
high_mmr=df.groupby("make")["mmr"].sum().sort_values(ascending=False).head(10)
```

```python
high_mmr.plot(kind="bar",x=high_mmr.index,y=high_mmr.values,color="red")
plt.title("Marketing value of different brands")
plt.xlabel("cars")
plt.ylabel("marketing price")
```

Insights:-
Ford brand has the highest marketing value followed by cheverolet

```python
#12
low_mmr=df.groupby("make")["mmr"].sum().sort_values(ascending=True).head(10)
```

```python
low_mmr.plot(kind="line",x=low_mmr.index,y=low_mmr.values,color="brown",lw=5)
plt.xticks(rotation=90)
plt.title("brands having lowest marketing price")
plt.ylabel("marketing price")
```

Insights:-
Dodge is the least Market priced car after ford truck
