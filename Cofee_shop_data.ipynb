```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
path="/content/Cofee_shop_sales.csv"
```

```python
df=pd.read_csv(path)
df
```

```python
#1
transaction_time=df["transaction_time"].value_counts().head(10)
transaction_time
```

```python
plt.title("Transaction Time Of Cofee Shop")
plt.style.use("dark_background")
plt.pie(transaction_time,labels=transaction_time.index,autopct="%1.1f%%")
```

Insights:-
9:31 a.m. is the time when most transactions take place in coffee shops


```python
#2
plt.title("product_category VS The price of that unit")
sns.lineplot(x="product_category",y="unit_price",data=df,errorbar=None,marker="*",color="red",markersize=15)
plt.xticks(rotation=90)
```

Insights:-
According to the graph, coffee beans have the highest unit price

```python
#3
plt.title("Total Bills In Months")
sns.barplot(x="Month Name",y="Total_Bill",data=df,estimator=np.max)
```

Insights:-
The total bill is highest in coffee shops during June, May, April, and January, while it is lowest during March and February

```python
#4
filter1=df[df["Month Name"]=="June"]
products=filter1["product_category"].value_counts()
products
```

```python
plt.figure(figsize=(10,6))
plt.title("Sales of June Month")
plt.pie(products,labels=products.index,autopct="%1.1f%%")
plt.legend(title="products.index",bbox_to_anchor=(1.5,1))
```

Insights:-
Coffee had the highest sales in the month of June 39.6%

```python
#5
plt.title("Total Bill Of Tea in March")
filter2=df[(df["product_category"]=="Tea") & (df["Month Name"]=="March")]
sns.lineplot(x="Day Name",y="Total_Bill",data=filter2,estimator=np.mean,errorbar=None,marker="*",markersize="15",markerfacecolor="yellow",color="cyan")
```

Insights:-
The total bill for tea in the month of March is highest on Monday

```python
#6
plt.figure(figsize=(10,6))
sns.barplot(x="product_category",y="Total_Bill",data=df,hue="Month Name",palette="rainbow")
plt.xticks(rotation=90)
plt.title("Month Wise Product Category Vs Totall Bill")
```

Insights:-
For any product,the highest total bill has come in the months of June and January

```python
#7
filter3=df[(df["Month Name"]=="May") & (df["product_category"]=="Coffee")]
time=filter3["transaction_time"].value_counts().head(10)
time
```

```python
plt.bar(time.index,time,color="red")
plt.xticks(rotation=90)
plt.xlabel("Time")
plt.ylabel("Transactions")
plt.title("Time Wise Sales of Cofee")
```

Insights:-
On all days the most sales are at 10:20 and 8:58

```python
#15
monthly_sales = df.groupby(df['Month Name']).size()
monthly_sales.sort_values()
```

```python
plt.title("MONTH WISE TRANSACTIONS")
plt.ylabel("TRANSACTION COUNT")
sns.lineplot(x=monthly_sales.index,y=monthly_sales.values,marker="*",markerfacecolor="red",markersize=10)
```

Insights:-
In the first 6 months of an year,it can be seen that most number of transactions was held in june followed by may

```python
#16
transaction_qty_dist = df['transaction_qty']
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(transaction_qty_dist, bins=8, color='LIGHTGREEN')
plt.title('Distribution of Transaction Quantities')
plt.xlabel('Transaction Quantity')
plt.ylabel('Frequency')
```

Insights:-
The histogram shows that most of the transactions involve 1 or 2 items, with a significant decline in frequency for higher quantities.

```python
#17
sales_by_day = df.groupby(df["Day Name"]).size()
sbd=sales_by_day.sort_values()
sbd
```

```python
colors=["violet","indigo","blue","green","yellow","orange","red"]
plt.bar(sbd.index,sbd.values,color=colors)
plt.title("SALES TRENDS BY DAY")
plt.xlabel("DAY")
plt.ylabel("COUNT")
plt.xticks(rotation=90)
```

Insights:-
The bar chart shows a fairly consistent number of transactions throughout the week,with a slight change on monday,thursday and fridays

```python
#18
hf=df[["transaction_qty","Total_Bill","unit_price"]]
hf.corr()
```

```python
#19
Total_sales_month = df.groupby('Month Name')['Total_Bill'].sum().reindex(['January', 'February', 'March', 'April', 'May', 'June'])
Total_sales_month
```

```python
plt.figure(figsize=(8, 6))
plt.plot(Total_sales_month, marker='o', color='blue',markersize=15,markerfacecolor="red")
plt.title('Total Sales Trend Over the Months')
plt.xlabel('Month')
plt.ylabel('Total Sales')
```

Insights:-
The line graph displays a clear upward trend in total sales from January to June.
